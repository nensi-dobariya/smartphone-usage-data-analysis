```python
import matplotlib.pyplot as plt
df = pd.read_csv("smartphone_usage_365_days 4 1.csv")
plt.figure(figsize=(10,6))
plt.scatter(df['screen_time_hours'],df['notifications'],color='pink',alpha=0.7,edgecolors='black',marker='*')
plt.xlabel("Screen Time (Hours)", fontsize=12)
plt.ylabel("Notifications", fontsize=12)
plt.title("Screen Time vs Notifications", fontsize=14, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df['day'], df['screen_time_hours'], color='green', marker='d', linewidth=2)
plt.xlabel("Day", fontsize=12)
plt.ylabel("Screen Time (Hours)", fontsize=12)
plt.title("Daily Screen Time Trend", fontsize=14, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()

```

```python
plt.figure(figsize=(10,5))
plt.plot(df['day'],df['social_media_hours'],color='blue',label='Social Media')
plt.plot(df['day'],df['gaming_hours'],color='yellow',label='Gaming')
plt.xlabel("Day")
plt.ylabel("Usage Hours")
plt.title("Social Media vs Gaming Usage")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# CSV load
df = pd.read_csv("smartphone_usage_365_days 4 1.csv")

# 3D figure
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')

# Data
x = np.arange(len(df))          # index as numeric x-axis
y = df['screen_time_hours']
z = df['notifications']

# 3D scatter
ax.scatter(x, y, z, c=y, cmap='viridis', s=50)

# Labels
ax.set_xlabel("Days")
ax.set_ylabel("Screen Time (Hours)")
ax.set_zlabel("Notifications")
ax.set_title("3D Smartphone Usage Analysis", fontweight='bold')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation

# Data load
df = pd.read_csv("smartphone_usage_365_days 4 1.csv")

x = np.arange(len(df))
y = df['screen_time_hours']
z = df['notifications']

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')

# Scatter with cap-type marker
sc = ax.scatter(x, y, z, marker='^', color='green', s=60)

ax.set_xlabel("Days")
ax.set_ylabel("Screen Time (Hours)")
ax.set_zlabel("Notifications")
ax.set_title("Moving 3D Smartphone Usage Chart", fontweight='bold')

# Animation function (rotation)
def rotate(angle):
    ax.view_init(elev=30, azim=angle)

# Animation
ani = FuncAnimation(fig, rotate, frames=360, interval=50)

plt.show()
```
